```python
def find_book_id(x):
    book = x[0]
    author = x[1]
    look_df = books.loc[books['title']==book]
    if look_df.empty:
        return None
    else:
        if len(look_df) > 1:
            look_df = look_df.loc[(look_df['title'] == book) & (look_df['author'] == author)]
            if look_df.empty:
                return None
            else:
                return look_df.to_dict(orient='records')[0]['id']
        else:
            return look_df.to_dict(orient='records')[0]['id']
```

```python
class Parser:
    def __init__(self):
        self.df = pd.DataFrame(columns=['book', 'text', 'cover'])
    
    async def parse(self, i):
        async with aiohttp.ClientSession() as session:
            async with session.get(f'https://bookmix.ru/quotes/index.phtml?all=yes&begin={i}&num_point=100&num_points=100') as response:
                html = await response.text()
                soup = BeautifulSoup(html)
                for card in soup.find_all('div', 'col-12'):
                    text = card.find('i')
                    if not text:
                        continue
                    text = text.text.encode('iso-8859-1').decode('utf-8')
                    book_desc = card.find('p', 'universal-blocks-description')
                    if not book_desc:
                        continue
                    book_desc = book_desc.find_all('a')
                    book = book_desc[0].text.encode('iso-8859-1').decode('utf-8')
                    #author = book_desc[1].text.encode('iso-8859-1').decode('utf-8')
                    new_df = new_df.append({
                            'book': book,
                            'text': quote
                        }, ignore_index=True)
    
    def start(self):
        loop = asyncio.get_event_loop()
        loop.run_until_complete(self.parse(100))

pars = Parser()
pars.start()
```
